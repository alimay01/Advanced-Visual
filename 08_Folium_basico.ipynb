```python
m = folium.Map(location = coordenadas_ucm_docu,
    zoom_start=14,
    tiles='Stamen Terrain'
)
tooltip = 'UCM docu'
folium.Marker(coordenadas_ucm_docu,
              popup = 'UCM',
              tooltip = tooltip).add_to(m)
```

```python
m
```

```python
m = folium.Map(location = coordenadas_ucm_docu,
    zoom_start=14
)
tooltip = 'UCM docu'
folium.Marker(coordenadas_ucm_docu,
              popup = 'UCM',
              tooltip = tooltip).add_to(m)
```

```python
m
```

```python
tooltip = "Click me!"

folium.Marker(
    [40.43847105761711, -3.7014854890675513],
    popup="<a href='https://www.ucm.es/estudios/grado-turismo' target = '_blank'>Web</a>", # target para sacar en otra pestaa
    tooltip=tooltip,
     icon=folium.Icon(color="red", icon="info-sign"), # color de icono
).add_to(m)
folium.Marker(
    [40.439889422283926, -3.711783660232115],
    popup="<b>UCM Turismo</b>",
    tooltip=tooltip,
    icon=folium.Icon(color="blue"),
).add_to(m)

m
```
