### Carga de datos

Este dataset tiene ya 7 años pero sigue siendo válido para nuestro propósito.

La principal ventaja es que tiene una variale de grupo (o "factor") creada en español (la variable "Tipo") para facilitar los gráficos por tipo de vehículo.

```python
df= pd.read_pickle("/content/drive/MyDrive/Colab Notebooks/datos_vehiculos_2016.pkl")
```

```python
df.shape
```

```python
df.describe(include = 'all')
```

Segmentación por tipo de combustible:

```python
count_classes = pd.value_counts(df['Tipo'], sort = True)
count_classes.plot(kind = 'bar', rot=0)
plt.title("Tipos de vehículos UK 2016")
plt.xlabel("Tipo")
plt.ylabel("Frecuencia");
```

Ver todos las distintas marcas de coche:

```python
count_classes = pd.value_counts(df['Manufacturer'], sort = True)
count_classes.plot(kind = 'barh')
plt.title("Fabricantes de vehículos UK 2016")
plt.xlabel("Número de modelos a la venta")
plt.ylabel("Fabricante");
```

```python
plt.scatter(df['EngineCapacity'],
            df['MetricCombined'])

plt.xlabel('Tamaño motor (cm3)')
plt.ylabel('Consumo combinado')
plt.show()
```

Representar cuantos registros hay de cada fabricante y su combustible
(es decir, segmentado por fabricante y combustible):

```python
para_treemap = df[["Manufacturer",
                   "Tipo"]].value_counts() # cuenta valores
```

```python
para_treemap
```

```python
para_tree2 = para_treemap.reset_index(name = "count") # genera un dataframe con una nueva columna con nombre
```

```python
para_tree2
```

```python
para_treemap.dtypes
```

```python
# plot it
squarify.plot(sizes = para_tree2['count'],         # el área es en función del count( debe ser numérica )
              label = para_tree2['Manufacturer'],
              alpha=.8 ) # transparencia del color
plt.axis('off')
plt.show()
```

```python
# plot it
squarify.plot(sizes = para_tree2['count'],
              label = para_tree2['Manufacturer'],
              alpha=.8 ,
              pad = True)
plt.axis('off')
plt.show()
```
