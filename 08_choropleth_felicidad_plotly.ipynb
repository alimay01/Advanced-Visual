```python
import pandas as pd
import geopandas as gpd
import plotly.express as px
import matplotlib.pyplot as plt
```

En la tabla se observan los nombres de las columnas previamente nombradas

```python
df = pd.read_csv("https://raw.githubusercontent.com/pedroconcejero/ucm_master_2024_PYTHON/main/Database%20indice%20felicidad%202019.csv")

df.head()
```

```python
for col in df.columns:
    print(col)
```

```python
fig = px.choropleth(df,
                   locationmode = 'country names',
                   locations = df['Country or region'],
                   color = df['Score'], color_continuous_scale='viridis_r')
fig.show()
```
